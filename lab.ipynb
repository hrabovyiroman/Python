# Середнє значення raisedhands

```python
data = np.array([
    [240, 'M', 'KW', 72, 98, 96, 168],
    [152, 'F', 'SaudiArabia', 100, 98, 40, 140],
    [135, 'F', 'Egypt', 70, 95, 70, 140],
    [136, 'F', 'W', 70, 95, 70, 140],
    [138, 'F', 'Jordan', 100, 95, 90, 190],
    [133, 'M', 'KW', 12, 0, 70, 82],
    [114, 'M', 'KW', 11, 0, 50, 61],
    [38, 'M', 'KW', 0, 0, 4, 4],
    [2, 'M', 'KW', 10, 0, 30, 40],
    [113, 'M', 'KW', 25, 0, 69, 94]])
raised_hands = data[:, 3].astype(int)
mean_raised_hands = np.mean(raised_hands)
print(f"Середнє значення raisedhands: {mean_raised_hands}" )
```

# національність з найбільшою кількістю представників


```python
nationalities = data[:, 1]
unique, counts = np.unique(nationalities, return_counts=True)
nationality_counts = dict(zip(unique, counts))
most_common_nationality = unique[np.argmax(counts)]
most_common_count = counts[np.argmax(counts)]

print(f"Національність з найбільшою кількістю представників: {most_common_nationality} ({most_common_count} осіб)")
```
