# Ex2 - Getting and Knowing your Data



### Step 1. Import the necessary libraries (pandas)

```python
import pandas as pd
```

### Step 2. Import the dataset from Teams (US_Crime_Rates_1960_2014.csv)

```python
dataset = pd.read_csv('US_Crime_Rates_1960_2014.csv', index_col='Year', sep = ';')
dataset
```

### Show the last 3 rows of the dataFrame

```python
dataset.tail(3)
```

### Step 3. Set the 'Year' column as index

```python
dataset.set_index('Year', inplace=True)
```

### Step 4. See the first 6 entries, 'Population' and 'Violent' columns only

```python
subset = dataset.loc[:, ['Population', 'Violent']]
subset.head(6)
```

### Step 6. What is the number of observations in the dataset?

```python
num_observations = dataset.shape[0]
num_observations
```

### Step 7. What is the number of columns in the dataset?

```python
columns_observations = dataset.shape[1]
columns_observations
```

### Step 8. Print the name of all the columns.

```python
dataset.columns
```

### Step 9. What is the data type of each column?

```python
data_types = dataset.dtypes
data_types
```

### Step 10. Summarize the DataFrame.

```python
summary = dataset.describe()
summary
```

### Step 11. What is the average population was in US in 1960s (1960-1969)?

```python
average_population_1960s = dataset.loc['1960':'1969', 'Population']
average_population_1960s = average_population_1960s.mean()
average_population_1960s
```

### Step 12. What was the most frequent type of crime in 2014?

```python
crime_2014 = dataset.loc['2014']
most_frequent_crime = crime_2014.idxmax()
most_frequent_crime

```

### Step 13. What was the number of Robbery crimes between 2000 and 2014, including both?

```python
robbery_crimes = dataset.loc['2000':'2014', 'Robbery']
total_robbery_crimes = robbery_crimes.sum()
total_robbery_crimes
```

### Step 14. What % of population was killed in Murder crimes in 2000?

```python
population_2000 = dataset.loc['2000', 'Population']
murder_2000 = dataset.loc['2000', 'Murder']
percentage_population_killed = (murder_2000 / population_2000) * 100
percentage_population_killed
```

### Step 15. By what percentage with respect to Population did crime increase in 2014 compared to 1960?

```python
population_1960 = dataset.loc['1960', 'Population']
population_2014 = dataset.loc['2014', 'Population']

total_crime_1960 = dataset.loc['1960', 'Total']
total_crime_2014 = dataset.loc['2014', 'Total']

percentage_increase = ((total_crime_2014 - total_crime_1960) / total_crime_1960) * 100

print(f"The percentage increase in crime with respect to population from 1960 to 2014 was: {percentage_increase:.2f}%")
```

### Step 16. What types of crime affected more than 1% of population in 2010? (suppose each crime occurs only once for each person)

```python
population_2010 = dataset.loc['2010', 'Population']

one_percent_population = 0.01 * population_2010

crime_types_more_than_1_percent = dataset.loc['2010', ['Murder', 'Forcible_Rape', 'Robbery', 'Aggravated_assault', 'Burglary', 'Larceny_Theft', 'Vehicle_Theft']]
crime_types_more_than_1_percent = crime_types_more_than_1_percent[crime_types_more_than_1_percent > one_percent_population]
crime_types_more_than_1_percent
```

### Step 17. In which year the Violent crime rate (as % of Total) was the lowest?

```python
dataset['Violent_Crime_Rate'] = (dataset['Violent'] / dataset['Total']) * 100

year_with_lowest_rate = dataset['Violent_Crime_Rate'].idxmin()
year_with_lowest_rate
```

### Step 18. In which year the forcible rape crime rate (as % of Pupolation) was the highest?

```python
dataset['Forcible_Rape_Rate'] = (dataset['Forcible_Rape'] / dataset['Population']) * 100

year_with_highest_rate = dataset['Forcible_Rape_Rate'].idxmax()
year_with_highest_rate
```
